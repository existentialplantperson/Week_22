```python
Problem 1: 
    
    Index of Highest Value
Given an array (or list) of integers, you need to find and return the index of the highest value in the array.

Example:

indexOfHighest([-1, 0, 5, 1, 5, 3, 4, 3]) => 2
```

```python
x = [-1, 0, 5, 1, 5, 3, 4, 3]
import numpy as np
```

```python
# define funmction that accepts array
def highest_value_index(input_array):
    
    max_value = np.max(input_array)
    
    index_max_value = input_array.index(max_value)
    return index_max_value
```

## Chrome Driver

```python
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
import time

#pip install webdriver-manager
```

```python
executable_path = {'executable_path':ChromeDriverManager().install()}

#when testing use headless=False
#you want to see where it gets stuck
browser = Browser('chrome', **executable_path, headless=False)
```

```python
all_quotes = []
url = 'http://quotes.toscrape.com/'
browser.visit(url)

#brings up the website
```

```python
#use beautiful soup simultaneously with splinter(selenuium)
for i in range(1,11):
    soup = bs(browser.html, 'html.parser')
    all_divs = soup.find_all('div', class_='quote')
    for div in all_divs:
        try:
            quote_text = div.find('span', class_='text').text
            print(quote_text)

            author = div.find('small', class_='author').text
            print(author)
            
            #multiple tags exist, so we need another loop
            tags = div.find('div', class_='tags').find_all('a')
            tags_list = []
            for tag in tags:
                tags_list.append(tag.text)
            
            quote_dict = {'quote':quote_text, 
                          'author': author, 
                          'tags':tags_list}
            
            all_quotes.append(quote_dict)
        except Exception as e:
            print(e)
            
    #look at the links, they each exist as a 'Next' button
    browser.links.find_by_partial_text('Next').click()
    time.sleep(1)
    
df = pd.DataFrame(all_quotes)
df
```
